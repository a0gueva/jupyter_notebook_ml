```python
# using car sales csv

car_sales["Price"] = car_sales["Price"].str.replace('.', "")
car_sales
```

```python
# remvoe last 2 zeros
car_sales["Price"] = car_sales["Price"].str[:-2]
```

```python
car_sales
```

```python
car_sales["Sales Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
## Get the cumulative sum of all cars day to day cumsum will go down the rows adding up the specified column values

car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales["Price"].dtype
```

```python
car_sales.plot(x="Sales Date", y="Total Sales");
```

```python
car_sales, car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
```

```python
heart_disease = pd.read_csv("heart-disease.csv")

heart_disease
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease["age"] > 50
```

```python
over_50
```

```python
over_50 = pd.DataFrame(heart_disease[heart_disease["age"] > 50])
```

```python
over_50
```

```python
## create a figure with pyplot 00 method

fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel = "Age", ylabel = "Cholesterol")

ax.legend(*scatter.legend_elements(), title="Target");


## add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle = "--");
```

```python
## subplots with two axes

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

ax0.set(title="Heart Disease and Chol", xlabel="Age", ylabel="Cholesterol")

ax0.legend(*scatter.legend_elements(), title="Target")

## Axes 1 setup
scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])

ax1.set(title="Heart Diseas and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")

ax1.legend(*scatter.legend_elements(), title="Target")

ax1.axhline(over_50["thalach"].mean(), linestyle = "--");

fig.suptitle("Heart Disease Analysis", fontsize="16", fontweight="bold")
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="Oranges")

ax0.set(title="Heart Disease and Chol", xlabel="Age", ylabel="Cholesterol")

ax0.set(xlim=[50, 80])

ax0.legend(*scatter.legend_elements(), title="Target")

## Axes 1 setup
scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="winter")

ax1.set(title="Heart Diseas and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")

ax1.set(ylim=[60, 200])

ax1.legend(*scatter.legend_elements(), title="Target")

ax1.axhline(over_50["thalach"].mean(), linestyle = "--");

fig.suptitle("Heart Disease Analysis", fontsize="16", fontweight="bold")
```
